```python

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/kvzha/Downloads/archive/Top_Highest_Openings.csv")
```

```python
# bar graph
distributor_counts = df["Distributor"].value_counts() # count the occurrences of distributors
plt.bar(distributor_counts.index, distributor_counts.values) # plot bar plot

# create labels
plt.xlabel("Distributor")
plt.ylabel("Count")
plt.title("Number of Movies per Distributor")
plt.xticks(rotation=90) # tilt x labels
plt.show() # display graph
```

```python
# pie chart
plt.pie(distributor_counts.values, labels=distributor_counts.index, startangle=90, autopct='%1.1f%%')
plt.title('Distribution of Movies per Distributor')
plt.legend(title='Distributor', loc='center left', bbox_to_anchor=(1, 0.5)) # create legend
plt.axis('equal')
plt.show()
```

```python
# scatter plot
sca_x = df["Opening"]
sca_y = df["Total Gross"]
plt.scatter(sca_x, sca_y, color="b", alpha=0.5)

# labels
plt.title('Opening vs Total Gross')
plt.xlabel('Opening')
plt.ylabel('Total Gross')

# display plot
plt.grid(True)
plt.show()
```
